```python
page = urlopen(url)
```

```python
# decodes to html
html_bytes = page.read()
html = html_bytes.decode("utf-8")
```

## elicits relevant data

### BeautifulSoup for main content

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(html_bytes, "html.parser")
```

```python
soup.text
```

```python
header = soup.find("div", class_ = "Bgc-w Bdstart-1g Bdend-1g Mih-50 Pt-7 Pb-14")
headers = header.find_all("span", class_ = "ya-q-full-text")
# header = soup.find("div", class_ = "Mstart-75 Mr-14 Pos-r")
```

```python
headers[0].text
```

```python
# finds itemprop name
str_header = str(header)
ind = str_header.find('itemprop="text">')
str_header[ind:]
```

```python
# apparently this is a thing!
header_text = header.get_text()
print(header_text)
```

```python
# prunes after 6*\n
end_ind = header_text.find("\n\n\n")
end_ind
# header_text = header_text[0:end_ind]
```
